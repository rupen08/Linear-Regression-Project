```python
# x = np.array([1, 2, 3, 4, 5])
# y = np.array([5,7,9,11,13])

## need y = 2x + 3

## so, m = 2 and b = 3
```

```python
def GD(x, y):
    m_start = b_start = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.07
    plt.scatter(x, y, marker = "+")
    
    for i in range(iterations):
        y_predict = (m_start * x) + b_start
        residual = (1/n) * sum([val**2 for val in (y-y_predict)])
        m_der = -(2/n) * sum(x* (y - y_predict))
        b_der = -(2/n) * sum(y - y_predict)
        
        
        ## deciding the learning rate
        
        m_start = m_start - learning_rate * m_der
        b_start = b_start - learning_rate * b_der
        
        ## print all 
        
        print (f" m {m_start} and b {b_start} and residual {residual} with iteration {i} ")
    plt.plot(x, y_predict)
        
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

## always define x and y in the same cell if using jupyter
```

```python
GD(x,y)
```
